## What is feature engineering

**Feature Engineering**  is the process of using domain knowledge of the data to create fetures or variables to use in ML 

**Feature Engineering**

**Fundamnetal**    - To make ML algorithm work 

**Key**            - To improve ML model performance 

**Time Consuming** - Big effort to preprocess variables in our dataset before we use them  

- It include 
   - Missing value imputation = filling missing values in variables 
   
   - Catagorical varaible encoding = encoding catagorical into numbers
   
   -  Variable transformation
   
   - Creating and extracting new faetures

# Course requirement


Welcome to the lecture on how to approach this course. In this article, I suggest some guidelines on how to approach this course, depending on your level.

Beginners
If you are just starting your journey in data science, if you havent worked with more than 1-2 datasets, I recommend that you follow the sections in the suggested order. Watch the videos first and then read the accompanying Jupyter notebooks, where you will find notes around the topic covered in the video, and the code to implement the techniques. Try the techniques in the recommended datasets, if necessary, run every cell of the Jupyter notebook to make sure you understand what the code is doing. 

Intermediate
If you have some experience as a data scientist, if you have already worked on 3-4 different datasets and done some variable pre-processing, I recommend you look at the sections and visit those lectures that are new to you. Check the Jupyter notebook for an explanation. If the concept is not clear, go ahead and look at the video. And then try the technique in the suggested dataset or on a different dataset you are interested in.  

Advanced
If you have already extensive experience as a data scientist and are here to increase your battery of techniques to engineer features, I recommend you browse the course curriculum first and identify the techniques that are new to you. Next, check the Jupyter notebooks for an explanation of the assumptions and limitations of the technique, and the code. Then go ahead and read the suggested additional material for a deeper understanding and more examples. Finally, try the new methods on the datasets you are currently working on.  

Enjoy the course!

Setting up your computer
These are the software and technical requirements to be able to go through the course succesfully. There are requirements that are a MUST, and some that are optional. Take your time to go through this list, and make sure you have everything you need to get started.



Python and packages
You need Python 3.6 or above and the numerical computing libraries NumPy, pandas, SciPy, Scikit-learn, category-encoders and Feature-engine for the course. Make sure you have all of these libraries installed with the following versions:


pandas=>0.24.2

numpy=>1.16.2

scikit-learn=>0.22.0

scipy=>1.2.1

category-encoders=>2.0.0

feature-engine=>1.0.0


Note: If you already have these libraries installed, test that the code runs before changing the versions to avoid unnecesary work.



Install python and the numerical computing packages (optional)
If you need to install Python from scratch, I recommend the Anaconda distribution which comes with all the numerical computing libraries (NumPy, Scipy, Matplotlib, Seaborn, Scikit-learn and Jupyter notebooks). To install Python:

Visit the anaconda website

Scroll down and download the python 3.6 version (or later)

Install it according to your operating system by following the instructions

Note: If you already have a regular Python 2/3 installation (not anaconda), you may risk conflicting version issues with Anaconda if the PYTHONPATH is not updated correctly. In this case, update the libraries using your current Python installation.



Install Feature-engine

To install feature-engine run:

pip install feature-engine



Install category-encoders
To install category-encoders run:

pip install category-encoders



Working with environments (optional)
If you want to create a separate environment with a python 3.6 installation and the libraries needed for the course, you can follow this instructions using anaconda:

In the anaconda prompt:

conda create -n mynewenv python=3.6 anaconda
activate mynewenv
conda install numpy
conda istall pandas
pip install feature-engine
pip install category-encoders
Note: this is just a subset of the required libraries, you need to install all the required ones into your new environment.



Adding a new environment to the Jupyter Notebook (optional)
To add the created environments to the Jupyter notebook follow this link. Briefly, if using conda, from anaconda prompt:

source activate mynewenv
python -m ipykernel install --user --name mynewenv --display-name "Python py36"
source activate other-env
python -m ipykernel install --user --name other-env --display-name "Python (other-env)"


That is all! You are now ready to get started!
